```python
import numpy as np
import matplotlib.pyplot as plt
```

### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

It can be done in an Object Oriented approach or a pyplot/MATLAB approach. The O.O. approach is generally recommended as it can handle more complicated plots. 

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure is the whole figure / the canvas. Axes is the part of the figure the plot is attached on. It can be specified with x,y,(z,...)-coordinates. There can be several axes on on a figure. Axis sets the scale, limits and generate ticks and ticklabels.
An artist is everything that is visible on the figure (figure, axes, axis, text, ect.)

### When plotting in Matplotlib, what is the expected input data type?

You should use a numpy.array or objects that can be transformed to a numpy.array with np.asarry

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4,4,100)            
y = x**2
```

```python
# creating a figure and axes
fig, ax = plt.subplots()

# Object Oriented approch
ax.plot(x, y, color='blue', label='Parabola',  linewidth=4, linestyle=':') 
ax.set_xlabel("Mastering Gauss Elemination", fontsize=12)
ax.set_ylabel("Perception of own cognitive ability", fontsize=12)
ax.set_title("Learning Process", fontsize=18)
ax.grid(color="gray")
ax.legend()
```

```python
# pyplot/MATLAB approach
plt.plot(x, y)

plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title("Title")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4), layout='constrained')

colors = random_data_y
ax1.scatter(random_data_x,random_data_y, s=7, marker="^", c=colors, cmap="viridis", label = "Data points")
ax1.set_xlim([-2, 2])
ax1.set_ylim([-2, 2])
ax1.set_xlabel("x-label", fontsize=12)
ax1.set_ylabel("y-label",  fontsize=12)
ax1.set_title("Title",  fontsize=16)
ax1.grid()
ax1.legend();

colors = ['orange', 'red', 'orange', 'yellow', 'green']
ax2.bar(names, values, color=colors)     
ax2.set_xlabel("Fruit", fontsize = 12)
ax2.set_ylabel("Number",  fontsize = 12)
ax2.set_title("Fruit inventory",  fontsize = 16)
```
